```python
import pandas as pd
data = pd.read_csv("BitcoinHeistData.csv")
```

```python
data.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python
features = ["year", "day", "looped", "weight", "income"]
for x in features:
    plt.boxplot(data[x])
    plt.ylabel(x)
    plt.show()
```

```python
for x in features:
    plt.hist(data[x], bins = 50)
    plt.ylabel(x)
    plt.show()
```

```python
sns.pairplot(data[features], hue="year" markers = [""])
```

```python
import itertools
features = list(itertools.product(features, features, features))
```

```python
for x, y, z in features:
    if(x!=y):
        if(y!=z):
            if(x!=z):
                fig = plt.figure()
                ax = fig.add_subplot(111, projection='3d')
                ax.scatter(data[x], data[y], data[z], c='r', marker='o')
                ax.set_xlabel(x)
                ax.set_ylabel(y)
                ax.set_zlabel(z)
                plt.show()
```
