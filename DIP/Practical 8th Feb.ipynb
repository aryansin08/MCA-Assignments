```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def getCos(time, freq):
    t = 2* np.pi * time * freq
    amp = np.cos(t)
    return amp
```

```python
time = np.arange(-5*np.pi,5*np.pi,0.1)
freq = 0.1
amp1 = getCos(time,freq)
 
plt.subplot(3,1,1)
plt.title("Freq : 0.1")
plt.plot(time,amp1)
amp2 = getCos(time,0.5)
 
plt.subplot(3,1,2)
plt.title("Freq : 0.5")
plt.plot(time,amp2)
amp3 = getCos(time,1)
 
plt.subplot(3,1,3)
plt.title("Freq : 1")
plt.plot(time,amp3)
 
plt.show()
```

```python
freq = np.fft.fftfreq(time.shape[-1])

sp1 = np.fft.fft(amp1)
fig = plt.figure(figsize=(10, 10))
plt.subplot(3,2,1,title="SP1 Real")
plt.plot(freq, sp1.real)
plt.subplot(3,2,2,title="SP1 Imaginary")
plt.plot(freq, sp1.imag)
 
sp2 = np.fft.fft(amp2)
plt.subplot(3,2,3,title="SP2 Real")
plt.plot(freq, sp2.real)
plt.subplot(3,2,4,title="SP2 Imaginary")
plt.plot(freq, sp2.imag)
 
sp3 = np.fft.fft(amp3)
plt.subplot(3,2,5,title="SP3 Real")
plt.plot(freq, sp3.real)
plt.subplot(3,2,6,title="SP3 Imaginary")
plt.plot(freq, sp3.imag)
 
plt.show()
```
