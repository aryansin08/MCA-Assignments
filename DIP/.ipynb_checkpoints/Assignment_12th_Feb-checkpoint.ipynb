```python
#1. Finding the phase angle
import numpy as np
import cv2
from matplotlib import pyplot as plt
from skimage import io

img = cv2.imread('image.png', cv2.IMREAD_GRAYSCALE)
imgplot = plt.imshow(img)
plt.title('Alan Turing')
plt.show()

def findPhase(img):
    dft = np.fft.fft2(img)
    dft_shift = np.fft.fftshift(dft)
    magnitude = np.abs(dft_shift)
    angle = np.angle(dft_shift)
    return magnitude, angle
```

```python
#Take input from user
inp = int(input("Enter some number: "))
```

```python
#Funciton to reconstruct image using Inverse FFT
def reconstruct(magnitude,angle):
    combined = np.multiply(magnitude, np.exp(1j*angle))
    fftx = np.fft.ifftshift(combined)
    ffty = np.fft.ifft2(fftx)
    imgCombined = np.abs(ffty)
    return imgCombined
```

```python
magnitude , angle = findPhase(img)
print(magnitude," ",angle)
```

```python
# Multiplying the magnitude by the user input
reconstructedImage = reconstruct(magnitude,angle*inp)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

imgplot = plt.imshow(reconstructedImage) # after inverse FFT
plt.show()
```
