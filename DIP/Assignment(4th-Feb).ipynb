```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('gray.png',0)
#127: threshold value
#255 : max threshold value
#thresh_binary : type of threshold
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img,127,200,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
ret,thresh6 = cv2.threshold(img,127,255,cv2.CAP_PROP_CONTRAST)

titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV','CAP_PROP_CONTRAST']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5, thresh6]

for i in range(7):
    plt.subplot(3,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
'''
print(img[0,:])
print(thresh1[0,:])
print(thresh3[0,:])
print(thresh4[0,:])
print(thresh5[0,:])
print(thresh6[0,:])
'''
```
